```python
from gtts import gTTS
from IPython.display import Audio, display

# Define the text you want to convert to audio.
text_to_convert = "Its very good hacakathon yaaaaar."

# Initialize the gTTS object.
tts = gTTS(text=text_to_convert, lang='en')

# Define the filename for saving the audio.
audio_filename = "converted_audio.mp3"
tts.save(audio_filename)
print("Audio has been saved as", audio_filename)

# Display and play the audio in the notebook.
display(Audio(audio_filename, autoplay=True))
```
